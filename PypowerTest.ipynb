```python
import numpy as np
import pypower
from pypower.api import case14, ppoption, runpf, printpf
import pandas as pd
```

```python
Ppb=(1,10,20,30,40,50,60,70,80)
Qpb=(0.40,3.95,7.90,11.86,15.81,19.76,23.71,27.67,31.62)
x=[0,1,2,3,4,5,6,7,8]
y=[1,2,3,4,5,6,9,10,11,12,13]
ppc = case14().copy()
ppopt = ppoption(PF_ALG=2)
```

```python
r = runpf(ppc, ppopt)
cn = 7 #columna que queremos obtener
columna = [fila[cn] for fila in r[0]['order']['int']['bus']]

df = pd.DataFrame()


df['V']=0

df['indice']=['bus0','bus1','bus2','bus3','bus4','bus5','bus6','bus7','bus8','bus9','bus10','bus11','bus12','bus13']
df.set_index('indice', inplace=True)

df.V=columna
```
